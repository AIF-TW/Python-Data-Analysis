## 資料視覺化小技巧

- ### 直方圖與箱型圖的比較


名稱    |   優點     |   缺點    
-------|------------|----------
直方圖  | 呈現資料完整分佈 | 不易比較多個分布差異
相形圖  | 可由圖形直接了解統計量與極端值 | 呈現資訊量過少, 圖形美觀度較差

---
- ### 散佈圖的陷阱

在探討兩個數值變項的關聯時，我們很常會直接使用散佈圖來直接觀察，但散佈圖可不可能有時也會誤導我們？

```python
# 載入套件
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

dat = pd.read_csv('../common/dataset/boston_house.csv')
dat.head()
```

- #### Pseudo-continuous variable

```python
plt.scatter(dat['CHAS'], dat['NOX'])
plt.xlabel('CHAS')
plt.ylabel('NOX')
plt.show()
```

```python
# 較好的做法是使用群組的箱型圖做比較
dat.boxplot('NOX', by='CHAS')
```

```python
plt.scatter(dat['TAX'], dat['RAD'])
plt.xlabel('TAX')
plt.ylabel('RAD')
plt.show()
```

```python
plt.hist(dat['TAX'], bins=20)
plt.show()
```

```python
# 調整透明度參數使資料分布較明顯

plt.scatter(dat['TAX'], dat['RAD'], alpha=0.1)
plt.xlabel('TAX')
plt.ylabel('RAD')
plt.show()
```

- ## 小結

在資料的視覺化上，其實還有很多需要注意的地方，未來大家在進行資料探索時請務必記得多方面交叉比較資訊是否一致。 另外，雖然很多圖形非常的絢麗美觀，但在資訊的呈現上其實基本的統計圖表並不輸給它們，在決定選用何種圖形前可以先思考一下目前想呈現的資訊為何，才能選擇到適合的圖形唷。
